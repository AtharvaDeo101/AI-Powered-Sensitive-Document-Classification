```python
import os
import random
import pandas as pd
from faker import Faker

fake = Faker()

# Define some sensitive and non-sensitive patterns
sensitive_keywords = ["PAN", "Aadhar", "bank_statement", "insurance", "tax", "payroll", "contract", "agreement"]
non_sensitive_keywords = ["meeting_notes", "project_plan", "recipe_book", "employee_directory", "vacation_pictures"]

# Helper function to generate random file metadata
def generate_file_metadata():
    # Randomly choose file type and extension
    file_types = ['pdf', 'xlsx', 'docx', 'txt', 'jpg']
    file_type = random.choice(file_types)
    
    # Randomly decide if the file is sensitive
    if random.choice([True, False]):
        filename = f"{random.choice(sensitive_keywords)}_{fake.date_this_year()}.{file_type}"
        sensitivity = "Sensitive"
    else:
        filename = f"{random.choice(non_sensitive_keywords)}_{fake.date_this_year()}.{file_type}"
        sensitivity = "Non-Sensitive"
    
    # Randomly assign file size (in bytes)
    file_size = random.randint(50000, 5000000)  # Size between 50 KB to 5 MB
    
    # Create random creation and modification dates
    creation_date = fake.date_this_year()
    last_modified = fake.date_this_year()
    
    # Generate random file path using fake
    file_path = fake.file_path(depth=3)  # Create a random directory structure
    
    return {
        "File Name": filename,
        "File Path": file_path,
        "File Type": file_type.upper(),
        "File Size (Bytes)": file_size,
        "Creation Date": creation_date,
        "Last Modified": last_modified,
        "Predicted Sensitivity": sensitivity
    }

# Generate a dataset of 1000 files
data = [generate_file_metadata() for _ in range(1000)]

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Save to CSV or database
df.to_csv("file_metadata.csv", index=False)

# Preview the first few rows
print(df.head())
```

```python
import pandas as pd
import random
import os

# Define parameters for dataset generation
num_samples = 1000  # Total number of rows in the dataset
sensitive_keywords = ["payroll", "tax", "insurance", "contract", "agreement", "aadhar", "bank"]
file_types = ['pdf', 'xlsx', 'docx', 'jpg', 'txt']
base_path = "C:/Atharva/"

# Function to generate a random file name
def generate_file_name(is_sensitive):
    if is_sensitive:
        keyword = random.choice(sensitive_keywords)
        return f"{keyword}_statement_{random.randint(2020, 2030)}_{random.randint(1, 12)}.{''.join(random.choices(file_types, k=1))}"
    else:
        return f"document_{random.randint(1, 1000)}_report.{''.join(random.choices(file_types, k=1))}"

# Generate the dataset
data = []
for _ in range(num_samples):
    # Randomly decide if the file is sensitive
    is_sensitive = random.random() < 0.3  # 30% sensitive files
    file_name = generate_file_name(is_sensitive)
    file_type = file_name.split('.')[-1].lower()
    file_size = random.randint(10000, 1000000)  # File size between 10KB and 1MB
    file_path = os.path.join(base_path, file_name).replace("\\", "/")
    
    # Add row to dataset
    data.append({
        'File Name': file_name,
        'File Type': file_type,
        'File Size (Bytes)': file_size,
        'File Path': file_path,
        'Predicted Sensitivity': 'Sensitive' if is_sensitive else 'Non-Sensitive'
    })

# Create a DataFrame
df = pd.DataFrame(data)

# Save the dataset to a CSV file
df.to_csv("new_metadata.csv", index=False)

print(f"Synthetic dataset with {len(df)} rows saved to 'file_metadata.csv'")
```
